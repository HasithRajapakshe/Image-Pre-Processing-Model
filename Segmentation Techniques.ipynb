```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('bd.jpg')

# Resize the image
resized_image = cv2.resize(image, (800, 600))

# Convert image to grayscale
gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(blurred_image)

# Apply binary thresholding for segmentation
_, binary_image = cv2.threshold(equalized_image, 150, 255, cv2.THRESH_BINARY)

# Perform segmentation based on thresholding
segmented_image = cv2.bitwise_and(resized_image, resized_image, mask=binary_image)


# Display the images using matplotlib
plt.figure(figsize=(16, 12))

# Original Image
plt.subplot(231)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Grayscale Image
plt.subplot(232)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')

# Blurred Image
plt.subplot(233)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')

# Equalized Image
plt.subplot(234)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

# Binary Image
plt.subplot(235)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')

# Segmented Image
plt.subplot(236)
plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
plt.title('Segmented Image')

plt.tight_layout()
plt.show()
```
